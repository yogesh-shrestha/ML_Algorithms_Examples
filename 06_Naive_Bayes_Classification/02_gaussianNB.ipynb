```python
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Load data
iris = datasets.load_iris()
X = iris['data'][:, 2:]
y = iris['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# fit model
model = GaussianNB()
model.fit(X_train, y_train)

# plot the decision boundary
fig, ax = plt.subplots(figsize=(10,6))
cmap_light = mcolors.ListedColormap(['peachpuff', 'lightblue','wheat'])
cmap_dark = mcolors.ListedColormap(['rebeccapurple', 'saddlebrown', 'teal'])
_ = ax.set(xlabel='Petal Length', ylabel='Petal Width', title='Naive Bayes Classification')
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.03), np.arange(y_min, y_max, 0.03))
X_new = np.c_[xx.ravel(), yy.ravel()]
Z = model.predict(X_new).reshape(xx.shape)

_ = ax.pcolormesh(xx, yy, Z, cmap=cmap_light, shading='auto')
_ = ax.scatter(X[:,0], X[:,1], c=y, cmap=cmap_dark)
```
